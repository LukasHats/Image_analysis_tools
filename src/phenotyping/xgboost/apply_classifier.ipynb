```python
label_mapping = pd.read_csv('/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/Phenotypes/models/xgboost/240613/phenotype_label_mapping.csv')
table = pd.read_csv(csv_path)
image_ID = csv_path.split('/')[-1].split('.')[0]
```

```python
table = pd.read_csv(csv_path)
image_ID = csv_path.split('/')[-1].split('.')[0]
table = table[table['area'] >= 4]
table.drop(columns=[" 1", " 2", " 3", " 4", " 5", " 6", "191Ir", "193Ir", "near_bone"], inplace=True)
table.rename(columns={'centroid-0': 'Y_centroid', 'centroid-1': 'X_centroid', 'Object': 'CellID'}, inplace=True)
table['image_ID'] = image_ID
table['index'] = table['image_ID'] + '_' + table['CellID'].astype(str)
table.set_index('index', inplace=True)
cols_to_add = table[['HistoneH3', 'CD98','X_centroid', 'Y_centroid', 'CellID', 'image_ID']]
table.drop(columns=['HistoneH3', 'CD98','X_centroid', 'Y_centroid', 'CellID', 'image_ID'], inplace=True)
transformed = np.arcsinh(table.iloc[:, 0:32])
table.drop(columns=table.columns[0:32], inplace=True)
phenotypes = pd.concat([transformed, table], axis=1)
```

```python
model = joblib.load('/Users/lukashat/Documents/PhD_Schapiro/Projects/Myeloma_Standal/QC/Phenotypes/models/xgboost/240613/best_xgb_model.joblib')
```

```python
predictions = model.predict(phenotypes)
phenotypes['predicted_label'] = predictions
phenotypes = phenotypes.reset_index()
phenotypes = pd.merge(phenotypes, label_mapping, left_on='predicted_label', right_on='EncodedLabel', how='left')
phenotypes.set_index('index', inplace=True)
result = pd.merge(phenotypes, cols_to_add, left_index=True, right_index=True)
```
